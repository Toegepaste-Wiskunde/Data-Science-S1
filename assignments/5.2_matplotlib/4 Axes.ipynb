## Axes
We've seen that `matplotlib` changes the axes to what it considers best for the current plot. However, sometimes it's better to set it yourself. Here we'll see how to control the Axes (scale, range, and ticks).

We'll again make use of the lines for sphere area and volume:

```python
## CREATE A FUNCTION
from numpy import pi

sphere_radius = [x/10 for x in range(0,50)]
sphere_area = [4*pi*x**2 for x in sphere_radius]
sphere_volume = [4/3*pi*x**3 for x in sphere_radius]
```

```python
import matplotlib.pyplot as plt
fig,ax = plt.subplots()
ax.grid(True)
ax.plot(sphere_radius,sphere_area, sphere_radius,sphere_volume)
fig
```

We can change the scale of an axis from the default linear to logarithmic. Then every tick indicates a multiplication by 10:

```python
ax.set_yscale('log')
fig
```

A logarithmic scale is mostly used for function that change exponentially. That's not the case for our functions so let's set it back to `linear`.

We can change the range of the x-axis that is show using the function `set_xlim`.

```python
ax.set_yscale('linear')

ax.set_xlim([0,2])
fig
```

### Subplots
The function `subplots` has the option to divide a figure in multiple subfigures. The number of subfigure rows is set using `nrows`, and the number of columns with `ncols`. When there are multiple subplots, the variable `axs` is not an Axes object, but **a list of Axes objects**.

```python
fig,axs = plt.subplots(nrows=2,ncols=1)
axs[0].plot(sphere_radius,sphere_area, 'r')
axs[0].grid(True)
axs[0].set_ylabel('area')
axs[1].plot(sphere_radius,sphere_volume, 'b')
axs[1].grid(True)
axs[1].set_xlabel('radius')
axs[1].set_ylabel('volume')
fig
```

Since both the above subfigures have the same x-range, we can set `sharex` to `True`, which means that the x-axis (range, markers, labels) is shared:

```python
fig,axs = plt.subplots(nrows=2,ncols=1,sharex=True)
axs[0].plot(sphere_radius,sphere_area, 'r')
axs[0].grid(True)
axs[0].set_ylabel('area')
axs[1].plot(sphere_radius,sphere_volume, 'b')
axs[1].grid(True)
axs[1].set_xlabel('radius')
axs[1].set_ylabel('volume')
axs[1].set_xticks([0,2.5,5])

fig
```
