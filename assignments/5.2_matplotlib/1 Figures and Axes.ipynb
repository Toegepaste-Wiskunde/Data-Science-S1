## Figures and Axes
Let's start by loading the library. It's commonly assigned to the variable `plt`:

```python
import matplotlib.pyplot as plt
```

Now the variable `plt` contains the functions that we use to make figures. To make a new figure we start with the command `subplots`:

```python
fig,ax = plt.subplots()
```

This function `subplots` returns two objects: `fig` and `ax`. The object `fig` is a Figure object that contains properties of the "board" on which we make our graphs. The graphs themselves are part of `ax` which is an Axis object. The Axis object also contains things such as labels or tick marks. 

If we want to see what is in our figure, we can just check the output of the variable `fig`:

```python
fig
```

Nothing is in our figure yet, since we haven't given a plotting command yet. We can adjust the width of the figure, with the option `figsize`:

```python
print(f'original width: {fig.get_figwidth()}')

fig,ax = plt.subplots(figsize=[10,5])

print(f'new width: {fig.get_figwidth()}')
fig
```

Most of the things we want to specify are part of the Axis of the figure. You can see that even without a graph, the figure already includes horizontal (x) and vertical (y) axis, that both range from 0 to 1 and have six "ticks".

The range and ticks of each axis are part of the Axis object and their values can be requested:

```python
fig,ax = plt.subplots()

print(f'Range x-axis: {ax.get_xlim()}')
print(f'Tick marks x-axis: {ax.get_xticks()}')

fig
```

Currently our figure is a bit boring, and there is not much more to say about it. 
So let's make it exciting by making our first graph:

```python
plotted_values = ax.scatter(x=[1,2,3,4],y=[3,2,4,1])
fig
```

The scatter plot show the data points at the requested positions. Note that `matplotlib` automatically updates the range of the axes to `[1,4]` to show all the data points. 

The scatter plot itself is assigned to the variable `plotted_values`, and this variable can be used to request information of the plot:

```python
plotted_values.get_offsets().data
```

An Axis object can contain multiple plots. Let's at another set of points:

```python
ax.scatter([1,3,4],[-1,2,3])
fig
```

Note that the plot range and tick marks are updated again:

```python
print(ax.get_xlim())
print(ax.get_xticks())
```

You'll have to look very closely to see what the actual values are in the scatter plot. It would be better if there were grid lines, so let's include those: 

```python
ax.grid(True)
fig
```

Much clearer! Note that in this notebook we updated the figure step-by-step, but normally you would make a figure in a single cell:

```python
fig,ax = plt.subplots()
ax.scatter(x=[1,2,3,4],y=[3,2,4,1])
ax.scatter(x=[1,3,4],y=[-1,2,3])
ax.grid()

fig
```

### Exporting the figure

Once you're done with editing your figure, you might want to save it as file (PNG,JPG,PDF), so that you can include it in a report. This can be done with the `savefig` option of the Figure object. This function requires a filename that specifies the location and file format (here: PNG). Optionally the resolution (and thus file size) can be added using the option `dpi` (dots-per-inch):

```python
from pathlib import Path # library for working with file paths

filename = Path.home() / 'my_first_graph.png'

fig.savefig( filename, dpi=300 ) 

print(f'SAVED FIGURE TO: {filename}')
```

Now you can find your figure in your home folder.

### Object-oriented vs direct approach

The above steps were shown in a "object-oriented" manner, in which the figure and axes are saved in variables that contain their properties. In this way you can keep multiple figures as variables and update them separately. 

However, you can also make graphs without assigning the Figure and Axes objects to variables. This is done by using the same functions as before, but with the Axis variable `ax` replaced with the variable that contains the library `plt`:

```python
plt.figure() # REPLACES fig,ax = plt.subplots()
plt.scatter(x=[1,2,3,4],y=[3,2,4,1])
plt.scatter(x=[1,3,4],y=[-1,2,3])
plt.grid()
plt.show() # REPLACES fig 
```

All graphs are drawn in the last Figure window that you activated. Using this direct approach can reduce the amount of code but also reduces the ability to update your graphs. The makers of `matplotlib` recommend to use the object-oriented approach.
