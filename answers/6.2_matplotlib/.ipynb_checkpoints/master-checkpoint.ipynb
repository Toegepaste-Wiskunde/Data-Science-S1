```python
import matplotlib.pyplot as plt
#%matplotlib inline
```

```python
fig,ax = plt.subplots()
```

```python
## CREATE A FUNCTION
xval = [x/100 for x in range(-100,100)]
yval = [x**3 for x in xval]
ax.plot(xval,yval)
fig
```
