## Styling

This section will discuss how to style your graphs with different kinds of lines, markers and colors. 

In order to have some data to show, we calculate the surface area and volume of a sphere with radius between 1 and 5 meters:

```python
## CALCULATE VALUES
from numpy import pi # the value of pi is part of the numpy library

sphere_radius = [x/10 for x in range(10,50)]
sphere_area = [4*pi*x**2 for x in sphere_radius]
sphere_volume = [4/3*pi*x**3 for x in sphere_radius]
```

In the previous section we used a scatter plot to show a small number of data points. In this section we can determine the area and volume for every radius using the formula. Thus plotting lines is more appropriate. This is done using the function `plot`. We can plot multiple lines with a single command by pairing x and y values:

```python
import matplotlib.pyplot as plt

fig,ax = plt.subplots()
ax.grid(True)
ax.plot(sphere_radius,sphere_area, # first (blue) line
        sphere_radius,sphere_volume) # second (orange) line
fig
```

Let's change the color of the lines. This can be done by including a format string for each line. Red is indicated with `'r'` and blue with `'b'`.

```python
fig,ax = plt.subplots()
ax.grid(True)
sphere_plot = ax.plot(sphere_radius,sphere_area, 'r', # r = red
        sphere_radius,sphere_volume, 'b') # b = blue
fig
```

We can also replace the line with markers or change it style to dash-dot, by changing the format string:

```python
fig,ax = plt.subplots()
ax.grid(True)
sphere_plot =  ax.plot(sphere_radius,sphere_area, 'r-.', 
        sphere_radius,sphere_volume, 'b+')
fig
```

The width of the plotted lines can be adjusted with the option `linewidth`:

```python
fig,ax = plt.subplots()
ax.grid(True)
sphere_plot =  ax.plot(sphere_radius,sphere_area, 'r-.', 
        sphere_radius,sphere_volume, 'b+',
        linewidth = 3 )
fig
```

Similarly, the marker size is controlled using option `markersize`:

```python
fig,ax = plt.subplots()
ax.grid(True)
sphere_plot =  ax.plot(sphere_radius,sphere_area, 'r-.', 
        sphere_radius,sphere_volume, 'b+',
        markersize = 3 )
fig
```

More information on colors, markers and line styles can be found in the [subsection on `plot`](https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.plot.html) of the `matplotlib` website.
