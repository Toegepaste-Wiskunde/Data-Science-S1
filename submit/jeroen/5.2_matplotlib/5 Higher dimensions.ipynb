## Higher dimensions
So far we've looked at simple plots with one independent variable on the x-axis and a dependent variable on the y-axis. But what if there is a function that is dependent on two variables? Here we'll show some examples for creating plots for data sets with three-dimensions. An overview with more examples can be found on the [`matplotlib` website](https://matplotlib.org/stable/plot_types/index.html).

We'll first make a data set. For this we use the `numpy` library that has functions to work with higher dimensional data:



```python
import numpy as np

# example taken from matplotlib
X, Y = np.meshgrid(np.linspace(-3, 3, 256), np.linspace(-3, 3, 256))
Z = (1 - X/2 + X**5 + Y**3) * np.exp(-X**2 - Y**2)
```

One way to show the third (`Z`) dimension, is by using color. This can be done using `imshow`:

```python
import matplotlib.pyplot as plt

fig,ax = plt.subplots()
Z_plot = ax.imshow(Z)
ax.set_xlabel('X') 
ax.set_ylabel('Y')
ax.set_title('imshow plot')
fig
```

However, we don't see which value matches with which color. For this we need to add a `colorbar`:

```python
fig.colorbar(Z_plot, label='value of Z')
fig
```

If we're more interested in the shape of the function, we can make a surface plot:

```python
fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
Z_plot = ax.plot_surface(X,Y,Z)
ax.set_xlabel('X') 
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('surface plot')
fig
```

The axes can be rotated to view the shape from different sides:

```python
ax.view_init(elev=10,azim=-60)
fig
```
