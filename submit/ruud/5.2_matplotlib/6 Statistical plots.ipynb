## Statistical plots
We've shown plots from small data sets or functions until now. Here we'll discuss a few methods to plot larger data sets. 

We'll first make a data set that consists of four sets of samples with different means and standard deviations. We'll make use of the `numpy` toolbox to create the samples.

```python
import matplotlib.pyplot as plt
import numpy as np
standard_deviations = [2,1,0,3]
means = [0,1,2,3]
sample_size = 100

xval = np.random.normal(means,standard_deviations,[sample_size,len(means)])
```

We'll first show the distribution of the samples in each set:

```python
fig, ax = plt.subplots()
for k in range(4):
    ax.plot([k]*sample_size,xval[:,k],'+')
ax.set_xticks([0,1,2,3])
ax.set_xlabel('set')
ax.set_ylabel('values')
ax.grid()

fig
```

We can make a histogram of one of the sets using the function `hist`:

```python
fig, ax = plt.subplots()

ax.hist(xval[...,0],bins=20)
fig
```

The mean and standard deviation of each set can be shown with the function `errorbars`:

```python
mean_xval = np.mean(xval,axis=0)
std_xval = np.std(xval,axis=0)

fig, ax = plt.subplots()
ax.errorbar([1,2,3,4],mean_xval,std_xval,
            fmt='+', linewidth=2, capsize=6)
ax.grid()
fig
```
