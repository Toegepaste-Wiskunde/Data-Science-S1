## Text
A good figure has axes with labels. And you can also add a title or at some text at a point of interest. Adding text is the topic of this section and we'll again use the formulas for the area and volume of a sphere:

```python
## CREATE A FUNCTION
import matplotlib.pyplot as plt
from numpy import pi

sphere_radius = [x/10 for x in range(0,50)]
sphere_area = [4*pi*x**2 for x in sphere_radius]
sphere_volume = [4/3*pi*x**3 for x in sphere_radius]

fig,ax = plt.subplots()
ax.grid(True)
ax.plot(sphere_radius,sphere_area)
fig
```

Let's start by labelling the axes:

```python
ax.set_xlabel('radius')
ax.set_ylabel('sphere area')
fig
```

It's also simple to add a title:

```python
ax.set_title('my first figure')
fig
```

If there are multiple graphs in a figure, it's recommended to add a legend that describes each line. This is done using the function `legend` of the Axis object:

```python
fig,ax = plt.subplots()
ax.grid(True)
ax.plot(sphere_radius,sphere_area,'r--',
        sphere_radius,sphere_volume,'kx'
        )
ax.set_xlabel('radius')
ax.legend(['area', 'volume']) # labels in the same order as plot
fig
```

What's still missing in the previous plot? Correct, the units of the different variables. The are $m$ for radius, $m^2$ for area, and $m^3$ for volume. We can get mathematical styled text by enclosing the text with symbols `$...$`:

```python
fig,ax = plt.subplots()
ax.grid(True)
ax.plot(sphere_radius,sphere_area,'r--',
        sphere_radius,sphere_volume,'kx'
        )
ax.set_xlabel('radius $[m]$')
ax.legend(['area $[m^2]$', 'volume $[m^3]$']) # labels in the same order as plot
fig
```

The function `annotate` of the Axis object is used to point out interesting data points:

```python
fig,ax = plt.subplots()
ax.grid(True)
ax.plot(sphere_radius,sphere_area,
        sphere_radius,sphere_volume)
ax.legend(['area', 'volume'])
ax.annotate(    text='intersection', # text to show
                xytext=[2,200], # where to put text
                arrowprops=dict(arrowstyle="->"), # type of arrow
                xy=[3,4*pi*3**2]) # point where arrow points toward
fig
```
